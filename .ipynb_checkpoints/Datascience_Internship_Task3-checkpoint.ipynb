```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest score
max_score = max(employees.values())

# Get all employees with the highest score
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"

# Convert to lowercase and split into words
words = query.lower().replace(",", "").replace(".", "").split()

# Count frequency
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

# Filter keywords appearing more than once
result = {word: count for word, count in freq.items() if count > 1}

print(result)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = [(i, val) for i, val in enumerate(sensor_readings) if val % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Count domains
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentages
total = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg_sales = sum(sales) / len(sales)

# Detect spikes (>30% above average)
for day, sale in enumerate(sales, start=1):
    if sale > avg_sales * 1.3:
        print(f"Day {day}: {sale}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences
id_count = {}
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

# Show duplicates
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} -> {count} times")
```
